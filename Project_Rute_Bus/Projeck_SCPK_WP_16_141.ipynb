## Projek Penentuan Rute Bus Sekolah Paling Efektif Berdasarkan Data Operasional Menggunakan Metode Weighted Product


### Latar Belakang
Bus sekolah di DKI Jakarta merupakan salah satu layanan transportasi publik yang ditujukan untuk memudahkan akses siswa menuju sekolah. Namun, tidak semua rute berjalan dengan efektif. Beberapa rute memiliki jumlah penumpang yang sangat tinggi, sementara rute lain sepi atau bahkan tidak beroperasi secara konsisten.

Dari data operasional yang tersedia (Januari–Mei 2014), terdapat 20 rute bus sekolah dengan data harian jumlah bus dan jumlah penumpang. Sayangnya, selama ini evaluasi rute masih bersifat manual dan belum menggunakan metode yang sistematis.

### Permasalahan
Dinas Perhubungan DKI Jakarta ingin menentukan rute bus sekolah mana yang paling efektif dari sudut pandang siswa sebagai pengguna utama, dengan mempertimbangkan beberapa faktor:
1. Seberapa banyak siswa yang menggunakan rute tersebut setiap hari?
2. Seberapa nyaman dan efisien penggunaan armada (tidak terlalu penuh, tidak terlalu kosong)?
3. Seberapa stabil jumlah penumpang dari hari ke hari?
4. Seberapa sering rute tersebut beroperasi?
5. Seberapa banyak bus yang tersedia di rute tersebut?

### Sumber Data
- Dataset: Data Operasi Bus dan Penumpang Bus Sekolah di DKI Jakarta
- Periode: Januari - Mei 2014
- Format: CSV
- Volume: ~1.000+ baris data
- Cakupan: 20 rute bus sekolah
