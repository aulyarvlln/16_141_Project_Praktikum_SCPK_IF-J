## Projek Penentuan Rute Bus Sekolah Paling Efektif Berdasarkan Data Operasional Menggunakan Metode Weighted Product


### Latar Belakang
Transportasi bus sekolah di DKI Jakarta merupakan salah satu fasilitas penting untuk mendukung mobilitas siswa. Dengan adanya 20+ rute yang beroperasi, diperlukan analisis untuk menentukan rute mana yang paling efektif berdasarkan data operasional nyata (jumlah penumpang, jumlah armada, dan konsistensi operasi).

Proyek ini menggunakan metode Weighted Product (WP) untuk memberikan rekomendasi berbasis data dalam pengambilan keputusan optimasi rute bus sekolah.

### Sumber Data
- Dataset: Data Operasi Bus dan Penumpang Bus Sekolah di DKI Jakarta
- Periode: Januari - Mei 2014
- Format: CSV
- Volume: ~1.000+ baris data
- Cakupan: 20 rute bus sekolah


Untuk Penamaan Criteria itu berdasar dibawah ini, karena dari data asli yg di dapat penjelasan criterianya kurang cukup sehingga kami memilih penamaan criteria nya sebagai berikut
